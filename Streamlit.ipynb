```python
%%writefile NLP_projectt.py
import streamlit as st
import pandas  as pd
from transformers import BartForConditionalGeneration, BartTokenizer
from transformers import MarianMTModel, MarianTokenizer
import os
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"

st.title('Summairze and Translated')
st.sidebar.header('User Input')

# upload_file = st.sidebar.file_uploader('Plase enter text ',type=['txt'])
# text_to_summarize = upload_file.read().decode("utf-8")

#Defination summariy model 
summarization_model_name = "facebook/bart-large-cnn"
summarization_tokenizer = BartTokenizer.from_pretrained(summarization_model_name)
summarization_model = BartForConditionalGeneration.from_pretrained(summarization_model_name)

#Defination translate model 
translation_model_name = "Helsinki-NLP/opus-mt-en-ar"
translation_tokenizer = MarianTokenizer.from_pretrained(translation_model_name)
translation_model = MarianMTModel.from_pretrained(translation_model_name)

# Function to summairize
def summarize_text(text):
    inputs = summarization_tokenizer(text, max_length=1000, return_tensors="pt", truncation=True)
    summary_tokens = summarization_model.generate(
        inputs["input_ids"], 
        max_length=130, 
        min_length=30, 
        length_penalty=2.0, 
        num_beams=4,  
        early_stopping=True
    )
    return summarization_tokenizer.decode(summary_tokens[0], skip_special_tokens=True)

# Function to translation
def translate_text(text):
    input_ids = translation_tokenizer(text, return_tensors="pt").input_ids
    generated_tokens = translation_model.generate(
        input_ids,
        max_length=130,
        num_beams=4,  
        early_stopping=True
    )
    return translation_tokenizer.decode(generated_tokens[0], skip_special_tokens=True)

upload_file = st.sidebar.file_uploader('Please upload a text file', type=['txt'])

if upload_file is not None:
# read content
    text_to_summarize = upload_file.read().decode("utf-8")
else:
# Enter Manually text  
    text_to_summarize = st.text_area("Enter text to summarize")
    
    if st.button('Summarize and Translate', key='summarize_translate_button'):
        summary = summarize_text(text_to_summarize)
        st.subheader("Summary")
        st.write(summary)
        
        translate = translate_text(summary)
        st.subheader("Translate to Arabic")
        st.write(translate)
    else:
        st.warning("Please enter some text to summarize.")
# Sidebar button
if st.sidebar.button('Summarize and Translate'):
    if text_to_summarize.strip():  # Check if the input is not empty
        summary = summarize_text(text_to_summarize)
        st.subheader("Summariy")
        st.write(summary)
        translate = translate_text(summary)
        st.subheader("Translate to Arabic")
        st.write(translate)
    else:
        st.warning("Please enter some text to summarize.")
```
