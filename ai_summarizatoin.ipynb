# Summarize Text

```python
from transformers import BartForConditionalGeneration, BartTokenizer

import pandas as pd

# load the Summarize model and tokenier
    summarization_model_name = "facebook/bart-large-cnn"
    tokenizer = BartTokenizer.from_pretrained(summarization_model_name)
    model = BartForConditionalGeneration.from_pretrained(summarization_model_name)

# Functoin Summarize text
def summarize_text(text):..
    inputs = tokenizer(text, max_length=1024 , return_tensors="pt", truncation=True)
    summary_tokens = model.generate(
        inputs["input_ids"], 
        max_length=130, 
        min_length=30, 
        num_beams=4, 
        early_stopping=True
    )
    return tokenizer.decode(summary_tokens[0], skip_special_tokens=True)

# load file in external file 
file_path_txt = r"C:\New folder\New Text Document.txt"

try:
    with open(file_path_txt, "r", encoding="utf-8") as file:
        text_to_summarize = file.read()
    print("Data loaded from text file.")

except FileNotFoundError:
    print(f"Text file not found at {file_path_txt}. Using default text.")
    
    text_to_summarize = """
    Dataset Description
    In response to the COVID-19 pandemic, the White House and a coalition of leading research 
    groups have prepared the COVID-19 Open Research Dataset (CORD-19). CORD-19 is a resource of 
    over 47,000 scholarly articles, including over 36,000 with full text, about COVID-19, SARS-CoV-2,
    and related coronaviruses.
    """

# text summary 
summary = summarize_text(text_to_summarize)
# print text
print("Summary:", summary)
```

# Translation Text

```python
from transformers import MarianMTModel, MarianTokenizer

# text =summary

# load the Translate model and tokenier
model_name = "Helsinki-NLP/opus-m t-en-ar"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Functoin Translate text
def translate_text(text):
    input_ids = tokenizer(text, return_tensors="pt").input_ids
    generated_tokens = model.generate(input_ids)
    translation = tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
    return translation

# text translate 
translate = translate_text(summary)

# print text
print("Translate:", translate)
```
